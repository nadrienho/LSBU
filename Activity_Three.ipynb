```python
days=[25,26,27,28,29]
snp_high=[6020.75,6025.42,6020.16,6029.65,6044.18]
snp_low=[5963.91,5992.27,5984.87,6003.98,6033.01]
FTSE_high=[8307.29,8285.66,8279.86,8030.47,8291.36]
FTSE_low=[8268.00,8246.22,8238.36,8271.99,8253.00]
```

```python
plt.plot(days,snp_high,color="green",marker='o',linestyle='solid')
plt.ylabel("Stock values")
plt.xlabel("Days")
plt.title("SnP highest stock values in November 2024" )
plt.show()
```

```python
plt.plot(days,FTSE_high,color="green",marker='o',linestyle='solid')
plt.ylabel("Stock values")
plt.xlabel("Days")
plt.title("FTSE highest stock values in November 2024" )
plt.show()
```

```python
plt.plot(days,snp_low,color="green",marker='o',linestyle='solid')
plt.ylabel("Stock values")
plt.xlabel("Days")
plt.title("SnP lowest daily stock values in November 2024" )
plt.show()
```

```python
plt.plot(days,FTSE_low,color="green",marker='o',linestyle='solid')
plt.ylabel("Stock values")
plt.xlabel("Days")
plt.title("FTSE lowest daily stock values in November 2024" )
plt.show()
```

```python
plt.plot(days,snp_high,label='SnP')  #plotting a multiple line charts
plt.plot(days,FTSE_high,label='FTSE')
plt.legend(loc=4)
plt.ylabel("Stock values")
plt.xlabel("Days")
plt.show()
```

```python
plt.plot(days,snp_low,label='SnP')  #plotting a multiple line charts
plt.plot(days,FTSE_low,label='FTSE')
plt.legend(loc=2)
plt.ylabel("Stock values")
plt.xlabel("Days")
plt.show()
```
